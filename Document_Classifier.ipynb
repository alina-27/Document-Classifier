```python
from sklearn.datasets import load_iris

iris = load_iris()

print("Dataset loaded successfully!")
print("Number of samples:", len(iris.data))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
documents = [
    "The football match was exciting",
    "The cricket team won the trophy",
    "Artificial Intelligence is changing technology",
    "Python is used for Machine Learning",
    "The election results were announced",
    "The government introduced a new policy"
]

labels = [
    "Sports",
    "Sports",
    "Technology",
    "Technology",
    "Politics",
    "Politics"
]
```

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(documents)
```

```python
model = MultinomialNB()
model.fit(X, labels)

print("Model trained successfully!")
```

```python
test = ["Python is the best programming language for AI"]

test_vector = vectorizer.transform(test)

prediction = model.predict(test_vector)

print("Predicted Category:", prediction[0])
```

```python
test = ["India won the cricket world cup"]

test_vector = vectorizer.transform(test)

prediction = model.predict(test_vector)

print("Predicted Category:", prediction[0])
```
